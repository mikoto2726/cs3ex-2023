```python
import pandas as pd
df = pd.read_csv('data/user_data.csv', sep=',', skiprows=0, header=0)
display(df.shape)
```

```python
print(df['gender'].value_counts())
```

```python
import matplotlib.pyplot as plt
plt.hist(df['age'], bins=20)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
