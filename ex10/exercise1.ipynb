```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
nd_c = np.zeros((10, 10, 3)).astype('uint8')
nd_c[1:3, 1:3, 2] = 255
nd_c[4:6, 4:6, 1] = 255
nd_c[7:9, 7:9, 0] = 255
```

```python
img_c = Image.fromarray(nd_c)
plt.imshow(img_c)
plt.show()
```

```python
nd_c = np.zeros((10, 10)).astype('uint8')
nd_c[2:7, 2:8] = 255
img_c = Image.fromarray(nd_c)
plt.imshow(img_c, cmap='gray')
plt.show()
```
