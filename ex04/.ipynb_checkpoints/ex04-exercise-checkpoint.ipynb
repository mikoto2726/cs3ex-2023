```python
import pandas as pd
csv_in ='Data/user_data.csv'
df = pd.read_csv(csv_in, sep=',', skiprows=0, header=0)
print(df.info())
print(df.head())
```

```python
display(df[df.duplicated(keep=False)])
```

```python
df=df.drop_duplicates().reset_index(drop=True)
print(df.shape)
```

```python
display(df)
```

```python
print(df.isna().sum())
```

```python
display(df)
```

```python
df['occupation']=df['occupation'].fillna("unknown")
```

```python
display(df)
```

```python
df=df.dropna().reset_index()
```

```python
df['age'] = df['age'].astype('int')
df['gender'] = df['gender'].astype('category')
```

```python
display(df)
```

```python
ser1 = 2023 - df['age']
df['birth_year'] = ser1 
```

```python
df = df.drop(columns='zip_code')
```

```python
display(df)
```

```python
print(df['occupation'].value_counts())
```

```python
df['occupation']=df['occupation'].replace('eductor','educatar')
```

```python
df['occupation']=df['occupation'].replace(['Writer','write'],'writer')
```

```python
display(df[(df['age'] < 10) | (70 <= df['age'])])
```

```python
df1=df[(df['occupation']=='artist') & (df['gender']=='M')]
display(df1['age'].mean())
```

```python
display(df)
```

```python
df = df[df['occupation'] != 'none']
df.reset_index(drop=True, inplace=True)
display(df)
```

```python
display(df[df['occupation'] == 'none'])
```

```python
df=df.sort_values(by='age', ascending=True)
display(df)
```

```python
df_max=df['age'].idxmax()
df_min=df['age'].idxmin()
print(df_max)
print(df_min)
```

```python
csv_out='new_user_data.csv'
df.to_csv(csv_out,index=False,encoding='utf-8-sig')
```
