```python
import numpy as np #numpyにnpという別名を付けてimport
import pandas as pd #pandasにpdという別名を付けてimport
import matplotlib.pyplot as plt #グラフ描画のためのライブラリmatplotlibをimportする

csv_in = 'Data/customer.csv'
df=pd.read_csv(csv_in, sep=',', skiprows=0, header=0)
display(df)
```

```python
display(df[df.duplicated(keep=False)])
```

```python
df2=df.drop_duplicates().reset_index()
```

```python
print(df2.shape)
```

```python
print(df2.isna().sum())
```

```python
df2['store']=df2['store'].fillna("Akabane")
```

```python
df3=df2.dropna().reset_index()
```

```python
print(df3.shape)
```

```python
df3['store']=df3['store'].replace(['hakusan','Hakusn'], 'Hakusan')
```

```python
csv_out='customer_cleaned.csv'
df3.to_csv(csv_out,index=False,encoding='utf-8-sig')
```

```python
df3['senior']=df3['age'].map(lambda x: 1 if x >= 60 else 0)
```
