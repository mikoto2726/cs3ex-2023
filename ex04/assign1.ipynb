```python
import numpy as np #numpyにnpという別名を付けてimport
import pandas as pd #pandasにpdという別名を付けてimport
import matplotlib.pyplot as plt #グラフ描画のためのライブラリmatplotlibをimportする

csv_in = 'Data/wheat.csv'
df=pd.read_csv(csv_in, sep=',', skiprows=0, header=0)
print(df.shape)
print(df.info())
```

```python
df_prod=df[df['Element']=='Production']
display(df_prod.head())
```

```python
df=df['Value'].mean()
display(df)
```

```python
df_prod_sort=df_prod.sort_values(by='Value', ascending=False).reset_index(drop=True)
display(df_prod_sort.head(10))
```

```python
display(df_prod_sort[df_prod_sort['Country']=='Japan'])
```

```python
df_prod_sort = df_prod_sort.drop(columns='Element')
display(df_prod_sort)
```

```python
csv_out='wheat_prod_sort.csv'
df_prod_sort.to_csv(csv_out,index=False,encoding='utf-8-sig')
```
