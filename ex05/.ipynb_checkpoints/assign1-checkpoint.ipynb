```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
df = sns.load_dataset('titanic')
print(df.shape)
print(df.info())
display(df.head())
```

```python
df = df.drop(columns='deck')
display(df.head())
```

```python
df = df.dropna()
display(df.head())
```

```python
print(df['survived'].mean())
```

```python
print(df['pclass'].value_counts())
```

```python
display(df.groupby('survived').mean())
```

```python
display(pd.crosstab(df['survived'],df['class']))
```

```python
display(pd.crosstab(df['survived'],df['adult_male'],normalize='columns'))
```

```python
display(df.pivot_table(index='survived',columns='sex',values=['fare'],aggfunc=np.mean))
```

```python
display(df.pivot_table(index='sex',columns='class',values=['survived'],aggfunc=np.sum))
```
