```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
csv_in = 'newly_confirmed_cases_daily.csv'
df = pd.read_csv(csv_in,sep=',',skiprows=0,header=0)
print(df.shape)
print(df.info())
display(df.head())
```

```python
df['Date'] = pd.to_datetime(df['Date'])
print(df.info())
```

```python
plt.plot(df['Date'], df['Tokyo'], label='Tokyo')
plt.plot(df['Date'], df['Hokkaido'], label='Hokkaido')
plt.xlabel('Date')
plt.ylabel('No. of positive cases')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
df = df.set_index('Date')
display(df.head())
```

```python
df_rolling7 = df.rolling(7).mean()
display(df_rolling7.head())
```

```python
start_date = pd.to_datetime('2021-05-01')
end_date = pd.to_datetime('2023-05-01')

plt.figure(figsize=(10,4))
plt.bar(df.index,df['ALL'],label='original')
plt.plot(df_rolling7.index, df_rolling7['ALL'],'--', label='7-day average',color='red')
plt.xticks(rotation=45,size=14)
plt.yticks(size=14)
plt.xlim(start_date,end_date)
plt.xlabel('Date',size=14)
plt.ylabel('No.of severe cases',size=14)
plt.legend()
plt.show()
```

```python
df_w_sum = df.resample('W').sum()
display(df_w_sum)
```

```python
display(df_w_sum[df_w_sum['ALL'] == df_w_sum['ALL'].max()])
```

```python
df['day_of_week'] = df.index.dayofweek
display(df.head())
```

```python
import datetime as dt
df_2022 = df[(df.index >= dt.datetime(2022,1,1)) & (df.index < dt.datetime(2023,1,1))]
display(df_2022)
```

```python
df2 = df_2022.groupby('day_of_week').sum()
display(df2)
print(df2)
```

```python
wday_list = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
plt.bar(wday_list, df2['Tokyo'])
plt.xlabel('Day of week',size=14)
plt.ylabel('Total number of positive cases',size=14)
plt.xticks(size=14)
plt.yticks(size=14)
plt.show()
```
