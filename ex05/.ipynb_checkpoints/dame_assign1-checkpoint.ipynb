```python
import numpy as np #numpyにnpという別名を付けてimport
import pandas as pd #pandasにpdという別名を付けてimport
import matplotlib.pyplot as plt #グラフ描画のためのライブラリmatplotlibをimportする
```

```python
import seaborn as sns
df=sns.load_dataset('titanic')
display(df)
```

```python
df=df.drop(columns='deck')
display(df)
```

```python
df = df.dropna()
display(df.head())
```

```python
print(df['survived'].mean())
```

```python
#display(df[df['pclass']==1].sum())
print(df['pclass'].value_counts())
```

```python
display(df.groupby('survived').mean())
```

```python
display(pd.crosstab(df['survived'],df['class']))
```

```python
display(pd.crosstab(df['survived'],df['adult_male'],normalize='all'))
```

```python
df.pivot_table(index='survived', columns='sex', values='fare')
```

```python
df.pivot_table(index='class', columns='sex', values='survived', aggfunc =sum)
```
