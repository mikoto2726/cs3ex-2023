```python
import numpy as np #numpyにnpという別名を付けてimport
import pandas as pd #pandasにpdという別名を付けてimport
import matplotlib.pyplot as plt #グラフ描画のためのライブラリmatplotlibをimportする
csv_in='data/7203.T.csv'
df_toyota=pd.read_csv(csv_in, sep=',', skiprows=0, header=0)
print(df_toyota.shape)
print(df_toyota.info())
print(df_toyota.head())
```

```python
csv_in='data/JPY=X.csv'
df_usd=pd.read_csv(csv_in, sep=',', skiprows=0, header=0)
print(df_usd.shape)
print(df_usd.info())
print(df_usd.head())
```

```python
df_merge=pd.merge(df_toyota, df_usd, how='inner', on='Date')
print(df_merge.shape)
print(df_merge.info())
display(df_merge.head())
```

```python
plt.scatter(df_merge['Adj Close_x'], df_merge['Adj Close_y'])
plt.xlabel('TOYOTA')
plt.ylabel('USD/JPY')
plt.show()
```
