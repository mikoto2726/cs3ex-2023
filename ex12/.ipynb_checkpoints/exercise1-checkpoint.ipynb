```python
import shutil, os, glob
```

```python
d = 'mult'
```

```python
if os.path.isdir(d):
    shutil.rmtree(d)
```

```python
os.makedirs(d)
for i in range(1,10):
    for j in range(1,10):
        file = '{}x{}.txt'.format(i,j)
        file_path = os.path.join(d, file)
        text = '{} x {} = {}'.format(i,j, i*j)  
        print(file_path)
        with open(file_path, 'w') as f:
            f.write(text)
```

```python
file_pat = '{}/*.txt'.format(d)
for file_path in glob.glob(file_pat):
    with open(file_path, 'r') as f:
        for line in f:
            print(file_path + " " +line)
```
