```python
import pandas as pd
csv_in='data/better_life_index.csv'
df=pd.read_csv(csv_in, skiprows=0, header=0)
print(df.shape)
print(df.info())
display(df.head())
```

```python
df_num=df.loc[:, "GDP per capita":"Feeling safe walking alone at night"]
display(df_num.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
df_corr=df_num.corr()
display(df_corr)
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(df_corr,annot=True,fmt='.2f',cmap="bwr")
plt.show()
```

```python
print(df_num.corrwith(df_num['GDP per capita']))
```

```python
plt.scatter(df_num['GDP per capita'],df_num['Life satisfaction'])
plt.xlabel("GDP per capita")
plt.ylabel('Life satisfaction')
plt.show()
```

```python
plt.scatter(df_num['GDP per capita'],df_num['Homicide rate'])
plt.xlabel("GDP per capita")
plt.ylabel('Homicide rate')
plt.show()
```

```python
import statsmodels.api as sm
X=df_num[['GDP per capita']]
y=df_num[['Life satisfaction']]
X_c=sm.add_constant(X)
display(X.head())
display(X_c.head())
```

```python
model=sm.OLS(y,X_c)
results=model.fit()
print(results.summary())
```

```python
print(results.rsquared_adj)
```

```python
print(results.params)
```

```python
import numpy as np
a0=results.params['const']
a1=results.params['GDP per capita']
x_min=X['GDP per capita'].min()
x_max=X['GDP per capita'].max()
x_min_max=np.array([x_min, x_max])
y_min_max=a0+a1*x_min_max
plt.scatter(X['GDP per capita'],y)
plt.plot(x_min_max, y_min_max, c='red')
plt.xlabel('GDP per capita')
plt.ylabel('Life satisfaction')
plt.show()
```

```python
x_test=np.array([50000])
y_test=a0+a1*x_test
print(y_test)
```
